## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    `vertices` should be a numpy array of integer points.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=8):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    # Define matrices for lane row vectors
    line_left = []
    line_right = []
    line_avg = np.zeros((2,4), int)
    
    # Separating left and right lane lines
    #  / For right lane lines endpoints are switched to fulfil identical indexing of bottom endpoint
    #  / (x1, y1) represent endpoint on bottom, while (x2,y2) represent endpoint towards top of the image
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2-y1)/(x2-x1)
            if (slope > 0.25) and (slope < 10):
                line_right.append([x2,y2,x1,y1])
            elif (slope < -0.25) and (slope > -10):
                line_left.append([x1,y1,x2,y2])
                
    # Averaging out left and right lane separately
    # / setting line_left & line_right as numpy arrays + defining intercept
    line_left = np.asarray(line_left)
    line_right = np.asarray(line_right)
    
    
    line_avg[0,:] = np.sum(line_left, axis=0) / line_left.shape[0]
    line_avg[1,:] = np.sum(line_right, axis=0) / line_right.shape[0]
    
    intercept_avg = np.zeros((2,1))
    slope_avg = np.zeros((2,1))
    
    # Extrapolating averaged line to bottom of image and
    for i in range (2):
        [x1, y1, x2, y2] = line_avg[i,:]
        slope_avg[i] = (y2-y1)/(x2-x1)
        intercept_avg[i] = y1 - slope_avg[i] * x1
        line_avg[i,0] = (img.shape[0] - intercept_avg[i]) / slope_avg[i]
        line_avg[i,1] = img.shape[0]
        
    # Extrapolating averaged line to highest detected line endpoint
    line_avg[0,3] = np.amin(line_left[:,3], axis=0)
    line_avg[0,2] = (line_avg[0,3] - intercept_avg[0]) / slope_avg[0]
    line_avg[1,3] = np.amin(line_right[:,3], axis=0)
    line_avg[1,2] = (line_avg[1,3] - intercept_avg[1]) / slope_avg[1]
    
    # Extrapolating averaged line based on fixed image position
    #line_avg[0,3] = 0.65 * img.shape[0]
    #line_avg[0,2] = (line_avg[0,3] - intercept_avg[0]) / slope_avg[0]
    #line_avg[1,3] = 0.65 * img.shape[0]
    #line_avg[1,2] = (line_avg[1,3] - intercept_avg[1]) / slope_avg[1]
    
    for x1,y1,x2,y2 in line_avg:
        img = cv2.line(img, (x1, y1), (x2, y2), color, thickness)

        
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    

    
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )


````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.

# Function for providing image visualizing masked area
def visualize_mask(img, vertices, frame_color):
    img_mask = cv2.cvtColor(img,cv2.COLOR_GRAY2RGB)
    img_mask = cv2.polylines(img_mask,vertices, True,frame_color, 3)
    return img_mask

# Implementation of lane detection
def lane_detect(image_init, kernel_size, canny_threshold_low, canny_threshold_high,\
                width_ratio, height_ratio, rho, theta,\
                threshold, min_line_length, max_line_gap, plots):

    # Read in image and transform color to grayscale
    image = grayscale(image_init)


    # Define kernel and use gaussian smoothing
    image_blur = gaussian_blur(image, kernel_size)


    # Define thresholds and use Canny filter 
    image_canny = canny(image_blur, canny_threshold_low, canny_threshold_high)

    # Apply mask for relevant line detection area using a quadrangle
    # Parameterized to symmetrically adjust height and width of upper side
    imshape = image_canny.shape
    vertices = np.array([[(0,imshape[0]),\
                          (math.floor(imshape[1]*(1-width_ratio)),math.floor(imshape[0]*(1-height_ratio))),\
                          (math.floor(imshape[1]*width_ratio), math.floor(imshape[0]*(1-height_ratio))),\
                          (imshape[1],imshape[0])]],\
                        dtype=np.int32)

    image_masked = region_of_interest(image_canny, vertices)

    # Hough transform parameters 
    # rho: distance resolution between lines within same grid point in Hough space
    # theta np.pi/180*1 #distance resolution between lines within same grid point in Hough space
    # threshold: number of lines within same grid point that identify a "common" line for representation
    # min_line_length: = minimum length of line to be considered
    # max_line_gap: maximum distance between adjacent lines to be treated as one line (while satisfying rho & theta)
    
    # Use houghlines on canny image and draw on blank image
    image_lines = hough_lines(image_masked, rho, theta, threshold, min_line_length, max_line_gap)
    
    image_blended = cv2.addWeighted(image_init,0.8,image_lines,1.0,0)

    
    ###############################################################################################
    # Plotting greyscale, smoothened and canny detected images for parameter optimization purpose #
    ###############################################################################################
    
    if plots == True:

        fig, axs = plt.subplots(1,3)
        fig.set_size_inches(15, 3)

        fig.add_subplot(1,3,1)
        axs[0].set_title('Greyscale')
        axs[0].axis('off')
        plt.imshow(visualize_mask(image,vertices,(0,255,0)), cmap='gray')

        fig.add_subplot(1,3,2)
        axs[1].set_title('Masked')
        axs[1].axis('off')
        plt.imshow(image_masked, cmap='gray')

        fig.add_subplot(1,3,3)
        axs[2].set_title('Hough lines')
        axs[2].axis('off')
        plt.imshow(image_blended, cmap='gray')
    
    return image_blended

### Setting the parameters for the pipeline ###

#Canny parameters
kernel_size = 5
canny_threshold_low = 90
canny_threshold_high = 120

#Mask parameters
width_ratio = 0.55

height_ratio = 0.4

#Hough transform parameters
rho = 1
theta = np.pi/180
threshold = 35
min_line_length = 10
max_line_gap = 2

### Setting the folders for input and output ###
dir = 'test_images'
dir_output = 'test_images_output'

pathlib.Path(dir_output).mkdir(parents=True, exist_ok=True)

list = os.listdir(dir)
number_files = len(list)
print('The following images are being processed:\n')
print(list)

### Setting whether to output plots
plots = True

### Calling the pipeline for all images ###

for file in list:
    image = mpimg.imread(dir + '/' + file)
    output = lane_detect(image, kernel_size, canny_threshold_low, canny_threshold_high,\
                    width_ratio, height_ratio, rho, theta,\
                    threshold, min_line_length, max_line_gap, plots)
    mpimg.imsave(dir_output + '/' + file, output, format='png')
````
